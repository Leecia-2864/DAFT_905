```python
#Step 1: call Github API to find out the cold joke contained in the 24 secret files in the following repo:
response = requests.get('https://api.github.com/search/code?q=.0001.scavengerhunt in:path+repo:ironhack-datalabs/scavenger');
search = response.json() 
search
```

```python
import json
import requests

import pandas as pd
data = pd.DataFrame(search)
data
#data.sort_values(items)
```

```python
def flatten(data, col_list):
    for column in col_list: #loops through the columns of the data
        flattened = pd.DataFrame(dict(data[column])).transpose()#flattens the dictiionary of colums into a table 
        columns = [str(col) for col in flattened.columns] #append colums of flattened in a list
        flattened.columns = [column + '_' + colname for colname in columns] # get the columns for the repo columns
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
nested_columns = ['items']

flat = flatten(data, nested_columns)


flat
```

```python
contents_list = []
headers = {'Accept': 'application/vnd.github.v3.raw', };
headers
```

```python
files = search['items'];
files = sorted(files, key=lambda file: file['name']);
for file in files:
    content_path = ('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}').format(file['path']);
    response = requests.get(content_path, headers=headers);
    contents_list.append(response.text);
```

```python
print(contents_list)
```

```python
clean_contents = [word.strip('\n') for word in contents_list]
joke = " ".join(clean_contents);
```
